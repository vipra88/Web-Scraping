```python
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page=2',headers=headers).text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
print(soup.prettify())
```

```python
company=soup.find_all('div',class_="companyCardWrapper")
```

```python
len(company)
```

```python
name=[]
reviews=[]
rating=[]
salaries=[]
details=[]

for i in company:
    
    name.append(i.find('h2').text.strip())
    reviews.append(i.find('a',class_="companyCardWrapper__ActionWrapper").text.strip())
    rating.append(i.find('div',class_="companyCardWrapper__companyRatingWrapper").text.strip())
    salaries.append(i.find('a',class_="companyCardWrapper__ActionWrapper").text.strip())
    details.append(i.find('div',class_="companyCardWrapper__tertiaryInformation").text.strip())
    
```
